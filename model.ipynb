```python
model=load('/content/model.joblib')
```

```python
def classify_image(model, img):
  img = cv2.imread(img)
  img = cv2.resize(img, (150, 150))
  cv2_imshow(img)
  img = img/255
  img = np.expand_dims(img, axis=0)
  cv2.waitKey(0)
  predictions = model.predict(img)
  if predictions[0][0] >= 0.5:
    result = 'PNEUMONIA POSITIVE'
  else:
    result = 'PNEUMONIA NEGATIVE'

  return result
  
```

```python
predict = classify_image(model, '/content/Chest-XR-LLL-pneumonia-2718974434.png')
print(predict)
```
